깊은 신경망이나 혹은 많은 은닉 유닛을 가진 분류기를 사용할 경우에는 데이터에 완전히 맞출 수 있지만 overfitting이 발생하여 모델의 일반화 기능이 좋지 않게 됩니다.

편향과 분산을 이해하기 위한 중요한 두 가지 요소는
1. Train set error 
>  훈련 세트 오차
2. Dev set error (validation)
> 개발 세트 오차

만약 고양이와 강아지 사진을 나눈다고 할 때 (고양이 = 1), (강아지 = 0)

### 사람의 오차율은 0에 수렴한다고 할 때
### 베이지안 최적 오차가 0에 가까울 때

#1

>Train set error = 1%

>Dev set error = 11%

이런 결과가 나온다면 Train set에 overfitting이 되어서 Dev set 일반화가 되지 않은 상태입니다. 

이런 경우 알고리즘이 높은 분산을 갖는다고 말합니다.



#2

그럼 다른 경우에는 어떻게 될까요?

>Train set error = 15%

>Dev set error = 16%

이 경우 이 알고리즘이 제대로 된 예측을 하지못하는 현상입니다.
훈련 데이터에도 잘 맞지 않는다면 underfitting 이 된 것입니다.

그리고 이 알고리즘은 높은 편향을 갖습니다.

하지만 Train error와 Dev error에 차이가 1% 이므로 일반화가 잘 되었다고 할 수 있습니다.


하지만 이 알고리즘은 Train set도 예측을 잘 못하는 등 높은 편향을 가지는 문제를 가지고 있습니다.

#3

###Dev오차가 더 큰 경우

>Train set error = 15%

>Dev set error = 30%

이 경우는 알고리즘이 높은 편향을 가지면서 높은 분산을 가지는 경우입니다.

Train에 대한 예측도 미흡하고 일반화도 제대로 이뤄지지 않은 상태입니다.

#4


###학습이 잘 이루어졌을 때

>Train set error = 0%

>Dev set error = 1%

편향과 분산이 낮은 모습을 보여줍니다.

만약 베이지안 최적 오차가 15% 정도가 됬다면 2번 분류기도 좋은 성능을 보인다고 할 수 있습니다.

분류기가 잘 작동하지 않을 경우 편향과 분산을 잘 분석해야합니다!

사진이 흐릿해서 사람도 구분하기 힘들다면 베이즈 오차가 더 커질 것이고 이 분석에 대한 세부 방식이 달라집니다!

정리하자면 Train set error을 통해 해당 분류기가 편향 문제가 있는지 알 수 가 있고 Dev set error와의 차이를 통해 분산 문제가 얼마나 나쁜지 알 수 있습니다.

> 위에 예시는 베이즈 오차가 작고(0에 가까움) Train set과 Dev set이 같은 확률 분포에서 왔다는 가정하에 이루어진 것입니다!

만약 다른 분포에서 왔다면 일반적으로 오차 차이가 날 수 밖에 없습니다!

그럴 경우 더 복잡한 분석을 통해 모델을 평가 해야합니다.
