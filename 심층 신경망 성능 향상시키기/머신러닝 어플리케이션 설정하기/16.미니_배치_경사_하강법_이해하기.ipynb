일반적인 배치 경사 하강법에서는 지속적으로 하락(올라가는 현상이 있어선 안됨!)

미니 배치에서는 새로운 데이터를 지속해서 쓰니, 일시적으로 증가 하락을 반복하지만 전체적으로 봤을 때 하강하는 모습을 보인다

- $X^{\{1\}} Y^{\{1\}}$는 쉬워서 비용이 낮지만 $X^{\{2\}} Y^{\{2\}}$는 더 어려운 미니 배치라서 잘못 표시된 샘플이 있다든지의 이유로 비용이 높아질 수 있습니다.

> 이것이 미니 배치 경사 하강법을 실행 했을 때 생기는 진동의 원인!

## 미니 배치 경사 하강법을 사용할 때 우리가 결정해야하는 하이퍼 피라미터

1. 미니 배치 사이즈

- 총 길이가 m일 때 미니 배치 사이즈를 m으로 설정하면 하나의 미니 배치밖에 가지지 못합니다.  > 배치 경사 하강법

- 다른 극단적인 예는 미니 배치 크기가 1일 경우입니다. > 확률적 경사 하강법이라고 불리는 알고리즘을 제공해줍니다.

> 1. 배치 경사하강법은 상대적으로 크게 움직입니다. 

> 2. 확률적 경사하강법은 대부분 전역 최솟값으로 가게되지만 어떤 경우는 잘못된 방향을 가르켜 잘못된 곳으로 가기도 합니다.
   - 극단적으로 노이즈가 많을 수 있지만 평균적으로 좋은 방향으로 이동하게됩니다.


보라색 = 확률적 경사 하강법

파란색 = 배치 경사 하강법

image.png

확률적 경사 하강법의 장점 > 

1. 하나의 샘플만 처리한 뒤 진행하여 매우 간단
2. 노이즈도 작은 학습률을 사용해 줄일 수 있다

확률적 경사 하강법의 단점 >

1. 벡터화에서 얻을 수 있는 속도 향상을 잃게된다는 점

2. 항상 최솟값에 수렴한다고 보장은 못함 
   - 하지만 더 일관도게 전역의 최솟값으로 이동


---
- 한 번에 하나의 훈련 세트를 진행하기에 각 샘플을 진행하는 방식이 매우 비효율적!

- 미니 배치 크기가 너무 크거나 작지도 않을 때 최고의 효율을 보인다.

- 적당한 미니 배치 사이즈는 많은 벡터화를 얻게되고, 미니배치 사이즈가 1000이라면 1000개의 샘플에 벡터화를 하게됨 > 한 번에 샘플을 진행하는 속도가 더 빨라짐


- 전체 데이터 셋이 진행되길 안 기다려도 됨


# 미니 배치 사용법!

1. 작은 훈련 세트라면 그냥 배치 경사 하강법을 사용! > 전체로 해도 빠르게 됨
    - 샘플이 2000개보다 적은 경우

2. 큰 데이터 셋이라면 64~512 값의 길이를 가지는 미니 배치가 가장 일반적
    - 컴퓨터 메모리의 접근 방식을 보면 2의 제곱인 것이 코드를 빠르게 실행시켜줌!


이전에 1000으로 설정했는데 1024으로 하는게 좋음!
