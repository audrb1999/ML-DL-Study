# ***9. 입력값의 정규화***

신경망의 훈련을 빠르게 할 수 있는 하나의 기법은 입력을 정규화 하는 것입니다.

다음과 같이 두 가지 입력 특성이 있는 훈련 세트가 있습니다.
x의 입력 특성은 2차원이며 현재 보이는 그림이 훈련세트의 산포도 입니다.

image.png

두 단계에 따라서 입력을 정규화 해보겠습니다.

1. 평균에서 빼는 것 즉 0을 만드는 것입니다.
  - $\mu = \frac 1 m$ * $\Sigma x^{(i)}$  -> i는 제곱이 아닌 몇 번째를 의미
  > i는 0~m까지

x = x - μ를 하여 0의 평균을 갖게 될 때 까지 훈련세트를 이동하는 것입니다. 

image.png

평균값들이 0에 분포

2. 분산을 정규화

feature x1이 x2 보다 더 큰 분산을 가지고 있습니다 (값이 넓게 분포)

 - $\sigma^2 = \frac 1 m \times \Sigma x^{(i)}**2$ -> 원소별 제곱

$\sigma^2$ 각 특정의 분산에 대한 벡터

이미 평균을 빼 버렸기 때문에 $x^{i}$의 요소별 제곱은 그냥 분산임

그 다음 각각의 샘플을 얻어 $\sigma^2로 나눠줍니다.$

> x /= $\sigma^2$

image.png

x1과 x2의 분산은 이제 모두 1과 같습니다.

이것을 훈련데이터를 확대하는데 사용한다면 테스트 세트를 정규화 할 때도 같은 σ, μ를 사용해야 합니다.

훈련 세트와 테스트 세트를 다르게 정규화하면 안돼기 때문입니다.
