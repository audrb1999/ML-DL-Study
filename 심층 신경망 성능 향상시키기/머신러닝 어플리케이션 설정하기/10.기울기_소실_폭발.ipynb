#  기울기 소실/ 기울기 폭발

> 심층 신경망을 훈련시킬 때 문제점은 경사의 소실과 폭발

매우 깊은 신경망을 훈련시킬 때, 미분값 혹은 기울기가 아주 작아지거나 커질 수 있습니다

image.png

이 그림처럼 매우 깊은 신경망을 훈련시키는 경우를 예로 들겠습니다.

각 층마다 w1, w2, w3,...wl 까지의 가중치를 가지고 있고 활성화 함수 g(z) = z를 사용합니다.

b[l]은 0 입니다.


g(z) = z 이고 b = 0 이기에


a = z입니다.

a[1] = g(z[1]) = z[1] = w[1] * x 단순히 가중치 곱한 값들을 곱하는 것

만약 여기서 가중치들이 w[1.5, 1.5] 값을 가진다면 지속적으로 1.5를 곱해 기하급수적으로 값이 커지게 될 것입니다. 1.5에 층의 개수의 제곱만큼


따라서 매우 깊은 신경망을 가진다면 기울기가 폭발하게 됩니다.

반대로 0.5값을 가진 행렬이라면 층이 깊어질 수 록 기울기가 거의 사라지게 될 것입니다.

 여기서 얻을 수 있는 것은 w의 값이 단위 행렬보다 조금 더 크다면 활성화값은 폭발하고, 반대로 적다면 활성값은 기하급수적으로 감소하게 됩니다.

마찬가지로 기울기도 기하급수적으로 증가하거나 감소할 수 있습니다.

하지만 마이크로소프트에서 L = 852개 (층이 852)개인 모델을 만들었습니다.

이런 깊은 신경망에서 활성값이나 경사가 L에 대한 함수로 기하급수적으로 증가하거나 감소한다면 활성값들이 크게 변동되게 됩니다.

특히 기울기가 작아지게 된다면 학습이 어려워지게 됩니다.
