# 지수 가중 이동 평균

경사 하강법보다 빠른 몇 가지 최적화 알고리즘이 존재! 

그 알고리즘들을 이해하기 위해서는 지수 가중 평균을 사용할 수 있어야 합니다.

통계학에서 지수 가중 이동평균이라고 부릅니다.

image.png

여기 일별 온도가 있습니다.

이 분포에서 지역 평균이나 이동 평균의 흐름을 계산하고 싶다면

$ \theta$ = 해당 날의 기온

V = 온도

$V^{[0]} = 0$

$V^{[1]} = V^{[0]} * 0.9 + 0.1 * \theta1$

$V^{[2]} = V^{[1]}* 0.9 + 0.1 * \theta2$

image.png

그러면 이렇게 일별 기온의 지수 가중 평균을 얻게됩니다.

$\beta$ = 0.9라고 하면

$V^t = V^{[t-1]} * \beta + (1-\beta ) * \theta1$

$V^{[t]}$를 계산하면 대략적으로 $\frac 1 {(1-\beta)} * day'avg$(일별 기온 평균)과 같습니다.

그래서 $\beta$가 0.9라면 10 * 일별 기온 평균 = 10일간 기온의 평균과 같습니다.

그것이 붉은 선입니다.

만약 $\beta$ 값을 0.98로 한다면

50 * 일별 기온 평균이므로 50일간 기온 평균과 같아지고 더 많은 날짜의 기온의 평균을 이용하기에 곡선에 노이즈가 적어집니다.

그러나 더큰 범위에서 평균 기온을 찾기에 올바른 값에서 더 멀어집니다.

그래서 기온이 바뀔 경우 지수가중평균 공식은 더 느리게 작용합니다.

$\beta$가 0.98이면 이전 값에 더 많은 가중치를 주고 현재의 기온에 작은 가중치를 주게 됩니다.

그래서 온도를 내리거나 올리면 이 선들(지수가중평균)은 $\beta$가 커서 더 느리게 작용합니다.

보시면 빨간색은 일찍 온도가 올라가고 내려가지만

초록색은 빨간색보다 느리게 반응하는 것을 볼 수가 있습니다.

대신에 범위가 커져 노이즈에 더 강해집니다.

$V^{[0]} = 0$

$V^{[1]} = V^{[0]} * 0.9 + 0.1 * \theta1$ 

$V^{[2]} = V^{[1]}* 0.9 + 0.1 * \theta2$

$V^{[1]}$는 $V^{[0]}$를 90퍼 $V^{[1]}$온도를 10퍼 포함하고 있습니다.

$V^{[2]}$는 $V^{[1]}$ 위에 값을 90퍼 포합 ($V^{[0]}$ 81퍼, $V^{[1]} 9퍼 포함합니다.$)

즉 이전 기온들을 포함하게 되는데 $\beta$가 0.9면 이전데이터를 위에처럼 10일 동안은 현재 온도에 영향을 줄 수 있습니다.

즉 이것은 데이터의 전체적인 움직임을 알 수 있다.
