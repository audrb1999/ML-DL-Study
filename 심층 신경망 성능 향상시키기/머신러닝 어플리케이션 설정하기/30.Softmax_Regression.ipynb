마지막 출력의 shape은 (4, 1)이며 모든 예측값의 합은 1이 되어야합니다.

이런 출력값을 얻기 위해서 소프트맥스층을 사용하는 것!

소프트맥스는

$z = wa + b$ 아핀 변환값을

$t = e^z$ 네이피어 수에 넣고

$a = \frac {e^z} {\Sigma t} = \frac {t} {\Sigma t}$ 이렇게 연산을 하는 것!

#Softmax 분류기 훈련시키기

소프트 맥스함수에 넣을 때 값이 크면 결과도 크게나옴 > 스코어 점수가 높을 수록 확률이 높게 나옵니다.

많은 멀티 클래스 분류를 할 때 모든 값을 소프트 맥스하는 것보다는 가장 점수가 큰 값에만 소프트맥스를 해주는게 속도가 빠르다

소프트 맥스함수는 클래스가 둘 이상인 경우의 로지스틱 회귀를 일반화 해둔 것입니다.

소프트맥스 분류에서 주로 사용하는 손실 함수는 j=1 부터 C(여기선 4)까지 $y_j * log(\hat{y}_j)$ 의 합의 음수값이다.

- 끝으로 소프트맥스 출력층이 있는 경우 경사하강법을 어떻게 구현하는지 알아보자.
- 마지막 출력층이 $z^{[L]}$을 계산하는데 (C, 1) 차원이 된다.
- 여기에다 소프트맥스 활성화 함수를 취해서 $a^{[L]}$ 또는 $\hat{y}$을 얻는 것이다.
- 그러면 그 값을 활용해서 손실 함수를 계산할 수 있다.
- 이전에 신경망의 정방향 전파를 손실 함수를 구하는데에 다룬 적이 있다.
- 역방향 전파나 경사하강법은 어떤가?
- 역방향 전파에서 초기화를 위한 핵심 단계 핵심이 되는 식은 마지막 층에서 $z^{[L]}$의 미분이 (4, 1) 벡터인 y^에서 (4, 1) 벡터인 y를 뺀 것과 같다는 것이다.
- 클래스가 4개일 때 모두 (4, 1) 벡터가 되는 것이다. 일반적으로는 (C, 1) 차원.
- 우리의 일반적인 정의를 빌리자면 여기서 $dz^{[L]}$은 비용 함수를 $z^{[L]}$에 대해 편미분한 것이다.
- 이런 식으로 시작해서 신경망 전체에 대한 미분을 역방향 전파로 구하는 것이다.

- 이제 두 클래스 중 하나가 아니라 C개의 다른 클래스 중 하나로 분류하는 학습 알고리즘을 구현할 수 있다.
