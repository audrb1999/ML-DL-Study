# 기울기의 수치 근사

역전파를 구현할 때 경사 검사라는 테스트가 있습니다.

올바르게 역전파를 구현했는지 확인하는데 도움을 줍니다.

$F(x) = x^3 $ 일 때 

1. 작은 거리를 움직인다 $\epsilon$(엡실론) = 0.01 이라면

x-ε, x+ε

image.png

$f(x + a) - f(x - a)$ = 삼각형의 높이 

삼각형의 밑변은 2a


- 중심 차분법 $$\frac{f(x + \epsilon) - f(x - \epsilon)} {2\epsilon}$$

- 수치 미분법 $$\frac{f(x + \epsilon) - f(x)} {\epsilon}$$ 

이 값이 $g'(x)$와 비슷해야함!

위의 값을 계산하면 3.0001 정도가 나옴!

$x^3$를 미분해도 기울기가 3이 나옴!

이것을 경사 하강법이랑 같이 사용한다면 약 2배 정도느려지지만 정확하는 점에서 더 의미가 있다.

첫 번째의 오차는 0.0001 > $O(\epsilon^2)$ 오차를 나타내줌, 엡실론 제곱이 오차

두 번째는 $O(\epsilon)$ 오차가 더 큼 ε만큼 오차 발생

두 번째는 한쪽만 사용하고

첫 번째는 양쪽의 차이를 사용!
