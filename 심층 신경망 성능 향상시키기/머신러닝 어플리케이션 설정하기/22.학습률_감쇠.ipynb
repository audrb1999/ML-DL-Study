# 학습률 감쇠

학습 알고리즘의 속도를 높이는 한 가지의 방법은 시간에 따라 학습률을 천천히 줄이는 것입니다.

최소값에 다가갈수록 학습률을 줄여줘야 안정적으로 안착할 수 있다!

> 학습률값이 클 땐 크게 움직일 수 있어 최소값에 빠르게 다가가지만 제대로 안착하지 못하고 주변을 배회(크게 움직여서 안착을 못하고, 값이 크니 노이즈도 크므로)

image.png

파란색은 학습률이 고정된 것이고 초록색은 학습률이 점점 줄어드는 것!


구현 방법

훈련 세트를 나눠서 n 번째 epoch마다

$\alpha = \frac 1 {(1+decay\_rate) * epoch\_num} \times \alpha_0$

</br>

decay_rate = 감쇠율!

초기 학습률을 곱해주는 형태

$decay_rate = 1$

$\alpha_0 = 0.2$일 때 epoch 마다 위에 식대로 $\alpha$를 계산해보면

image.png

점점 감소하는 모습을 볼 수 있습니다.

여기서 중요한 건

- $\alpha_0$, 감쇠율같은 하이퍼 피라미터값들을 다양하게 시도하고 최적의 값을 찾는 것!


## 다른 방법들

1. 지수적 감소
   - $\alpha = 0.95^{epoch_num} \times \alpha_0$
   - 기하급수적으로 빠르게 학습률이 감소됩니다.

2. $$\alpha = \frac k {\sqrt {epoch_num}} \times \alpha_0$$

3. $$\alpha = \frac k {\sqrt {t}} \times \alpha_0$$

4. 이산 계단
   - 일정 시간, 반복마다 학습률을 반씩 줄이는 기법

k = 상수

t = 미니 배치의 갯수

학습률 감쇠는 최적값에 도달하게 해주는 기능을 해주지만 앤드류 교수님은 우선순위를 낮게 잡아두심!
