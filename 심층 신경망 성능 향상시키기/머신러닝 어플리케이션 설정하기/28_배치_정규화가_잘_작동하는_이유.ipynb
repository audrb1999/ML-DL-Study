# 배치 정규화가 잘 작동하는 이유는 무엇일까요?

1. 이유

입력 특성 x를 평균 0 분산 1로 정규화하는 것이 학습속도를 올려줌, 어떤 특성은 0\~1까지 어떤 특성은 1\~1000인 것을 입력 특성 x에 대해 비슷한 범위를 갖도록 정규화하여 학습 속도를 높이는 것!

2. 이유

10번째 층이 1번 층처럼 앞쪽층의 가중치의 변화에 영향을 덜 받기 때문

배치 정규화가 잘 되는 이유중 하나는 이전 층의 가중치 영향을 덜 받게 하는데 있습니다. 은닉층 값의 분포 변화를 줄여줘서, 입력 값의 분포를 제한하기 때문입니다. 즉, 배치 정규화는 입력값이 바뀌어서 발생하는 문제를 안정화 시킵니다. 앞층과 뒷층의 매개변수의 상관 관계를 줄여주기 때문에, 학습속도를 향상시킬 수 있습니다.

왼쪽 데이터에 대해서 분류를 학습을 했기에 오른쪽 데이터에 대해서는 제대로 분류를하지 못한다.

편향된 데이터로 학습해서는 초록색 모델을 만들지 못한다.

이처럼 데이터 분포가 변화하는 것을 공변량 변화라고한다.

> 검은 고양이 데이터가 유색 고양이 데이터로 변화!

데이터가 분포가 바뀐다면 학습 알고리즘을 다시 학습해야합니다.

하지만 모든 데이터가 동일할 수 없기에 지속적으로 입력값이 변하면, 이전층 출력 값이 계속 바뀌기 때문에 공변량 변화를 느끼기는 마찬가지지만 정규화를 하면 값의 변동이 적어집니다. 

배치 정규화를 하면 값의 변동이 생겨도 평균과 분산값이 일정하게 유지가 됩니다.

즉 적어도 평균과 분산은 0과 1처럼 유지가 됩니다.

굳이 0과 1이 아니더라도 $\beta \gamma$로 대처가 가능합니다.

앞선 층에서의 매개변수가 바뀌었을 때 다음층에서 학습하게 될 값의 분포를 제한하는 것입니다.

배치 정규화는 앞선층에서 값의 변동을 더욱 안정화시키는 것이지요

입력값의 분포가 조금 바뀌더라도 조금만 바뀌는 것이됩니다.

앞선 층에서 계속 학습하면서 값을 바꾸더라도 뒤쪽층에서 겪는 부담을 줄이는 것

즉 각 층의 독립성을 높이고 이전, 다음층의 연결이 약화되는 것

따라서 신경망의 각 층이 스스로 배울 수있게 되죠

3. 규제효과

미니 배치에서 계산한 평균과 분산은 모든 데이터에서 계산한 평균 분산에 비해 노이즈가 있습니다.

미니 배치라는 작은 데이터 묶음에서 전체 데이터를 예측하는 것이기에 그렇습니다.

그래서 각층의 틸다 드롭아웃에도 노이즈가 생깁니다.

> 일반화 효과

하지만 드롭아웃, 배치 정규화의 노이즈는 작기 때문에 일반화도 작습니다.

즉 두개를 같이 사용할 시 일반화가 더 강해집니다.


큰 미니 배치를 사용하면 잡음이 줄기에 일반화가 줄어듭니다.

그리고 큰 미니 배치에 배치 정규화를 해도 잡음이 줄어드는 특성이 있습니다.

그래서 앤드류 응 교수님은 배치 정규화를 일반화 목적으로 사용하지 않습니다.
