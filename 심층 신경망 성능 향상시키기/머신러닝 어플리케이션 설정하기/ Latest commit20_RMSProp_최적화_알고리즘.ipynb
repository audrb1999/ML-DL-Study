# RMSProp 최적화 알고리즘

모멘텀을 사용하느 것이 경사 하강법을 빠르게 할 수 있다는 것을 배웠습니다.

RMSprop = root maen square prop이라고 불리는 알고리즘입니다.

이 알고리즘 역시 경사 하강법을 빠르게 해줍니다.

이전의 경사 하강법은 수직 방향으로 큰 진동이 있다는 것을 알 수 있습니다.

여기서 수직 축은 매개변수 b, 수평축은 매개변수 w라고 가정합니다.

여기서 우리가 할 일은 b방향의 속도는 낮추고 w 방향의 속도를 높여야 하는 것입니다.

여기서 지수 가정 평균은 s_dw라고 나타내겠습니다.

$s_dw = \beta  \space \times s\_dw \space + (1-\beta) \times dw^2 $

- 여기서 $dw^2$ 는 요소별 제곱, 이것은 도함수의 제곱을 지수 가중 평균 하는 것입니다!

s_dw도 위와 동일합니다.

위에서 구한 값으로 RMSprop는

- $$w := w - \alpha \frac {dw} {\sqrt {s\_dw}} $$

- $$b := b - \alpha \frac {db} {\sqrt {s\_db}} $$

이것으로 어떻게 b방향으로는 천천히 w방향으로는 빠르게 이동할 수 있을까요?

- s_dw값은 작아서 이 값으로 나누면 값이 커지고 s_db는 값이 커서 나누면 작아지게함!

image.png



- b 값이 위 아래로 진동하는 이유 = 빼주는 값이 커서 진동하는 것!
  - 빼주는 값을 줄여주는 방법을 통해 진동을 작게함


- w 값이 느리게 진행되는 이유 = 빼주는 값이 작아서
   - 빼주는 값을 크게해서 갱신을 크게함 (학습률을 높인 것과 동일한 효과!)

일반적으로 dw보다 db값이 더 큰데 그 이유는 수직 방향 b에서 기울기가 더 가파르기 때문입니다.

그래서 $db^2$는 상대적으로 크고 $dw$가 작기에 $dw^2$는 상대적으로 작습니다.

이게 제곱 평균 제곱근 줄여서 RMSprop입니다.

도합수를 제곱해서 제곱근을 얻기 때문입니다.

이제 모멘텀과 RMSprop를 합칠 것인데 햇갈리지않게 

- 모멘텀의 $\beta = \beta1$

- RMSprop의 $\beta = \beta2$

라고 표기하겠습니다.

알고리즘이 0으로 나눠지지 않도록 주의하는 것도 필요합니다.

만약 s_dw값이 0에 매우 가깝다면 기울기가 폭발할 수 있습니다.

그러므로 실제로 구현할 때 수학적 안정성이 보장되야 합니다.


$\epsilon$ = 나누는 값을 말하는 것

$\epsilon = 10^{-8}$이면 합리적인 기본값이지만 이보다 약간 큰 값도 수학적 안정성을 보장합니다.

너무 작은 값을 사용하지 말자!
