# 단일 실수 평가 기준

하이퍼파라미터를 조정하거나 여러 학습 알고리즘을 시도해보고 서로 다른 머신러닝 시스템을 만들면서 실수 평가 기준이 있다면 훨씬 과정이 빨라질 겁니다.

두 분류기의 성능을 평가하는 합리적인 방법 중 하나는 정밀도와 재현률을 보는 것입니다

- 정밀도 = 분류기가 고양이로 인식한 예시 중 진짜 고양이가 몇 퍼센트냐
$\frac {TP} {TN + TP}$ 

- 재현률 = 실제 고양이 사진 중 분류기가 몇 퍼센트 정확하게 인식했는가
$\frac {TP} {TP + NP}$

정밀도와 재현율은 종종 트레이드 오프 관계에 엤다고 알려져 있습니다.

Precision = 정밀도
Recall = 재현률

다음과 같이 정밀도가 높은 모델 > 임계값이 낮다 > 데이터 포함 범위가 크다

재현률 > 얼마나 정확하게 인식했냐 > 임계값 값이 매우 높은 모델 즉 애매한 거 다 거르고 정확한 것만 한 것 > 일반화가 떨어짐

이 두 모델중 어떤게 더 좋은지 모른다. > 상황에 따라 달라진다

만약 아동 유튜브 필터라면 정밀도를 높인 것이 좋다 > 애매한 것을 다 거를 때 유해물이 아닌 것도 걸러지지만 아동 유튜브에 유해물이 뜨는 것보단 더 좋음 

우리는 idea > code > experiment(실험)을 계속 반복하면서 저 두 가지 지표로 모델들을 비교하여 좋은 모델을 고르려고한다면 저 지표보다는 F1 Score 지표를 사용하는 것이 좋다

F1 Score = 정밀도와 재현률이 결합한 새로운 평가 기준

대략적으로 생각한다면 정밀도와 재현률의 평균으로 생각해도 됨

실제 F1 지수의 정의는 $ 2 / \frac 1 P + \frac 1 R = 2 / \frac {R+P} {PR} = 2 \frac {RP} {R+P}$

> 수학에서는 정밀도 P와 재현률 R의 조화평균이라고 부릅니다.
