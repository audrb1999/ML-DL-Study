### 순차 탐색(Sequential Search)



문제: 어떤 수 x가 n개의 수로 구성된 리스트 S에 존재하는가?

- 해답: x가 존재하면 x의 인덱스가 주어지고 존재하지 않으면 0이 해답이 되야함!

- 파라미터 정의: 정수n (n>0 길이!), 리스트 S(인덱스 범위는 1~n까지), 원소 x(input)

- 입력 사례: S = [0, 10, 7, 11, 5, 13, 8], n =6, x=5

> 해답: location = 4, 원소의 인덱스 값을 반환



동작 방식

1. S의 첫째 원소에서 시작하여 x를 찾을 때까지(x가 없다면 끝까지)

2. 각 원소를 차례로 x와 비교한다.

3. 만약 x를찾으면 x의 인덱스를 리턴하고

4. x를 찾지 못하면 0을 리턴한다.

```python
# 메모리가 덜 쓰임
def seqsearch(n, S, x):
    location = 1
    while (location <= n and S[location] != x):
        location += 1 # x가 나올 때 까지 계속 원소를 찾음
    if (location > n): # 만약 없다면
        location = 0 # 0 반환
    return location # 있으면 그 원소 자리 반환

# 내 방식
def other(S, x):
    for a, i in enumerate(S):
        if i == x:
            return a
    return 0 
        
```

```python
S = [0, 10, 7, 11, 5, 13, 8]
x = 4
location = seqsearch(len(S) - 1, S, x)
print(location)

other1 = other(S, x)
print(other1)
```

### 리스트 (배열) 원소의 합 구하기

- 문제: n 개의 원소를 가진 리스트(배열) S의 원소의 합을 구하시오

- 해답: 리스트 S의 모든 원소들의 합

- 파라미터: 리스트:S, 정수:n

- 알고리즘: S의 모든 원소를 차례로 sum에 더하는 절차

1. sum을 0으로 초기화
2. 모든 S의 원소에 대해서 sum += S[i]를 실행
3. sum 리턴

```python
# 내 코드
def sum1(S):
    total = 0
    for i in S:
        total += i
    return total

def sum2(n, S):
    result = 0
    for i in range(1, n+1):
        result = result + S[i]
    return result
```

```python
S = [0, 10, 7, 11, 5, 13, 8]

suma = sum1(S)
print(suma)

sumc = sum2(len(S)-1, S)
print(sumc)
```

### 리스트의 정렬 문제(sorting)
> 정렬은 다양한 방법이 있습니다.

- 문제: n개의 수로 구성된 리스트 S를 오름차순으로 정렬하시오

- 해답: S 오름차순으로 정렬

image.png



알고리즘이 많다!

```python
#교환 정렬

def exchange(S):
    n = len(S)
    for i in range(n-1):# 값 한 개를 고르고
        # i + 1뒤에는 완벽하게 정리되어서 상관안해도 됩니다. 
        for j in range(i + 1, n): # 나머지 값과 비교해서 작은 값이 있으면
            if (S[i] > S[j]): # 두 개의 위치 변경
                S[i], S[j] = S[j], S[i]
    return S
```

```python
S = [-1, 10, 7, 11, 5, 13, 8]

x = exchange(S)
print(x)
```

### 행렬의 곱셈 문제

정방행렬 n x n 행렬의 곱을 구하시오

image.png

```python
def matrixmult (a, b):
    n = len(a)
    c = [[0]*n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            for k in range(n):
                c[i][j] += a[i][k] * b[k][j]
    return c
```

```python
a = [[2, 3], [4, 1]]
b = [[5, 7], [6, 8]]

c = [[0, 0], [0, 0]]


c = matrixmult(a, b)
print(c)
```
